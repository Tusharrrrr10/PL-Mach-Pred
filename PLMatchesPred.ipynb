```python
data1 = data.copy()
data1.head()
```

```python
data1.drop(['time','comp','round','venue','attendance','match report','notes','referee','captain'],axis=1,inplace=True)
```

```python
data1['dist'].fillna(data1['dist'].median(),inplace=True)
```

```python
data1["date"] = pd.to_datetime(data1["date"])
```

```python
data1.formation.value_counts()
```

```python
data1.formation = data1.formation.str.replace('◆', '')
data1.formation = data1.formation.replace('-0','')
```

```python
data1.formation.value_counts()
```

```python
data1.season.value_counts()
```

```python
data2 = data1.columns
data2 = data2.drop('season')
data2
```

```python
data1.duplicated(subset=data2).sum()
```

```python
data1.drop_duplicates(subset=data2, inplace=True)
```

```python
data1.duplicated(subset=data2).sum()
```

```python
def correct_seasons(data1):
    def get_correct_season(date):
        if date.month >= 8:
            return date.year + 1
        else:
            return date.year

    data1['season'] = data1['date'].apply(get_correct_season)
    return data1
```

```python
data1 = correct_seasons(data1)
data1['season'].value_counts().sort_index()
```

```python
data1.result.replace({'L':-1,'W':1,'D':0},inplace=True)
```

```python
data1.head()
```

```python
data1 = data1.assign(  day = data1.date.dt.day,
                        month = data1.date.dt.month, 
                        year = data1.date.dt.year )
```

```python
data1.drop('date',axis=1,inplace=True)
```

```python
data1.head()
```

```python
data1.formation.nunique()
```

```python
data1['formation'] = data1['formation'].astype('category')
```

```python
data1.dtypes
```

```python
#Label Encoder

```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data1["opponent"]=le.fit_transform(data1["opponent"])
data1["team"]=le.fit_transform(data1["team"])
# data1["result"]=le.fit_transform(data1["result"])
data1["formation"]=le.fit_transform(data1["formation"])
```

```python
data1.head()
```

```python
# X = data1.drop(['result'],axis=1)
# Y = data1.result

X = data1.drop(columns=['result']).values  # Features
Y = data1['result'].values
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)
 
print("Classification report: ")
 
print(classification_report(Y_test,Y_pred))
 
acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
#DT
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier(random_state=10,criterion='gini',splitter='best',min_samples_leaf=2,min_samples_split=3,max_depth=5)


model_dt.fit(X_train,Y_train)
Y_pred = model_dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)
 
print("Classification report: ")
 
print(classification_report(Y_test,Y_pred))
 
acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
model_dt.score(X_train,Y_train)
```

```python
classifier=DecisionTreeClassifier(random_state=10,criterion='gini',splitter='best',min_samples_leaf=2,min_samples_split=3,max_depth=5)
 
#performing kfold_cross_validation
from sklearn.model_selection import KFold
kfold_cv=KFold(n_splits=10,random_state=10,shuffle=True)
#print(kfold_cv)
 
from sklearn.model_selection import cross_val_score
#running the model using scoring metric as accuracy
kfold_cv_result=cross_val_score(estimator=classifier,X=X_train,
                                                 y=Y_train, cv=kfold_cv)
print(kfold_cv_result)
#finding the mean
print(kfold_cv_result.mean())
```

```python
#model tuning
#perform only when CV can be used for tuning
for train_value, test_value in kfold_cv.split(X_train):
    classifier.fit(X_train[train_value], Y_train[train_value]).predict(X_train[test_value])
    
Y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)
 
print("Classification report: ")
 
print(classification_report(Y_test,Y_pred))
 
acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
